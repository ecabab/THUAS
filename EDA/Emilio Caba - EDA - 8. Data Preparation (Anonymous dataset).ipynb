```python
import pandas as pd
import numpy as np
```

```python
df_20 = pd.read_csv('Data/df_20.csv')
df_40 = pd.read_csv('Data/df_40.csv')
df_60 = pd.read_csv('Data/df_60.csv')
df_80 = pd.read_csv('Data/df_80.csv')
df_100 = pd.read_csv('Data/df_100.csv')
df_120 = pd.read_csv('Data/df_120.csv')
```

```python
df_weather = pd.read_csv('Data/KNMI_Voorschoten_20181024_20190802.csv')
```

```python
df_20 = df_20.set_index('date')
df_40 = df_40.set_index('date')
df_60 = df_60.set_index('date')
df_80 = df_80.set_index('date')
df_100 = df_100.set_index('date')
df_120 = df_120.set_index('date')

df_20.index = pd.to_datetime(df_20.index)
df_40.index = pd.to_datetime(df_40.index)
df_60.index = pd.to_datetime(df_60.index)
df_80.index = pd.to_datetime(df_80.index)
df_100.index = pd.to_datetime(df_100.index)
df_120.index = pd.to_datetime(df_120.index)

df_weather.set_index('Date_and_time',inplace=True)
df_weather.index = pd.to_datetime(df_weather.index)
df_weather.index.names = ['date']
df_weather = df_weather.apply(pd.to_numeric)
```

```python
df_20_15min = df_20.groupby(pd.Grouper(freq='15T')).sum()
df_40_15min = df_40.groupby(pd.Grouper(freq='15T')).sum()
df_60_15min = df_60.groupby(pd.Grouper(freq='15T')).sum()
df_80_15min = df_80.groupby(pd.Grouper(freq='15T')).sum()
df_100_15min = df_100.groupby(pd.Grouper(freq='15T')).sum()
df_120_15min = df_120.groupby(pd.Grouper(freq='15T')).sum()
```

```python
for df in [df_20_15min, df_40_15min, df_60_15min, df_80_15min, df_100_15min, df_120_15min]:
    print(df.shape)
```

```python
df_15min = df_20_15min.join(df_40_15min, how='outer')
df_15min = df_15min.join(df_60_15min, how='outer')
df_15min = df_15min.join(df_80_15min, how='outer')
df_15min = df_15min.join(df_100_15min, how='outer')
df_15min = df_15min.join(df_120_15min, how='outer')
```

```python
df_15min.shape
```

```python
df_15min.dropna(inplace=True)
```

```python
df_deli = df_15min.drop(columns=df_15min.columns[::2])
df_cons = df_15min[df_15min.columns[::2]]
```

```python
df_deli.head(1)
```

```python
df_weather.head()
```

```python
df_deli = pd.concat([df_deli, df_weather], axis=1)
df_deli.shape
```

```python
df_cons = pd.concat([df_cons, df_weather], axis=1)
df_cons.shape
```

```python
df_deli = df_deli.dropna(subset=df_deli.columns[:-4])
df_cons = df_cons.dropna(subset=df_cons.columns[:-4])
```

```python
df_deli.fillna(method='ffill', inplace=True)
df_cons.fillna(method='ffill', inplace=True)
```

```python
df_deli.index.month
```

```python
df_deli['hour'] = df_deli.index.hour
df_deli['day'] = df_deli.index.day
df_deli['week'] = df_deli.index.week
df_deli['month'] = df_deli.index.month

df_cons['hour'] = df_cons.index.hour
df_cons['day'] = df_cons.index.day
df_cons['week'] = df_cons.index.week
df_cons['month'] = df_cons.index.month
```

```python
df_deli.head()
```

```python
df_cons.head()
```

```python
# Setting different columns (delivery, consumption, houses_info, dummy_variables)
df = pd.DataFrame(columns=['delivery','consumption', 'T', 'SQ', 'Q', 'N', 'hour', 'day', 'week', 'month'])

num_houses_cols = len(df_deli.columns) - 8
# Preparing the delivery and consumption arrays
ar_deli = np.array([])
ar_cons = np.array([])
for col in range(num_houses_cols):
    ar_deli = np.append(ar_deli, df_deli.values[:, col])
    ar_cons = np.append(ar_cons, df_cons.values[:, col])

# Inserting the data to the dataframe
df['delivery'] = pd.Series(ar_deli)
df['consumption'] = pd.Series(ar_cons)

df['T'] = df_deli['T'].values.tolist() * num_houses_cols
df['SQ'] = df_deli['SQ'].values.tolist() * num_houses_cols
df['Q'] = df_deli['Q'].values.tolist() * num_houses_cols
df['N'] = df_deli['N'].values.tolist() * num_houses_cols
df['hour'] = df_deli['hour'].values.tolist() * num_houses_cols
df['day'] = df_deli['day'].values.tolist() * num_houses_cols
df['week'] = df_deli['week'].values.tolist() * num_houses_cols
df['month'] = df_deli['month'].values.tolist() * num_houses_cols

display(df.head())
display(df.shape)
```

```python
df.to_csv('Data/FZ_models.csv')
```
